```python
#applying extra functions to create a new column og height category
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x <165 else ('Average' if x <180 else 'Tall'))
bios[['name', 'height_category']]
```

```python
#use if instead of lambda function
def categorize_athlete(row):
    if row['height_cm'] < 165 and row['weight_kg'] <65:
        return 'LightWeight'
    elif row['height_cm'] < 180 and row['weight_kg'] < 75:
        return 'Middleweight'
    else:
        return 'HeavyWeight'
    
bios['Category'] = bios.apply(categorize_athlete, axis = 1)
```
